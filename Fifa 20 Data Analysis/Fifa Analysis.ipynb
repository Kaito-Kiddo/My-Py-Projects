#### 4- Scatterpolar plot to compare a player's grothw over time
let's load the other datasets players from 2016 to 2019

```python
df_16 = pd.read_csv("players_16.csv")
df_17 = pd.read_csv("players_17.csv")
df_18 = pd.read_csv("players_18.csv")
df_19 = pd.read_csv("players_19.csv")
```

```python
print(list(df.columns))
```

player attributes column names

```python
attributes = ['pace','shooting','passing','dribbling','defending','physic','overall'] 
```

Creating a method to compare a Players growth over Time

```python
def playergrowth(name):
    data20 = df[df.short_name.str.startswith(name)]
    data19 = df_19[df_19.short_name.str.startswith(name)]
    data18 = df_18[df_18.short_name.str.startswith(name)]
    data17 = df_17[df_17.short_name.str.startswith(name)]
    data16 = df_16[df_16.short_name.str.startswith(name)]

    trace0 = go.Scatterpolar(
        r = [data20[attr].values[0] for attr in attributes],
        theta=attributes,
        fill = 'toself',
        name = '2020'
    )

    trace1 = go.Scatterpolar(
        r = [data19[attr].values[0] for attr in attributes],
        theta=attributes,
        fill = 'toself',
        name = '2019'
    )

    trace2 = go.Scatterpolar(
        r = [data18[attr].values[0] for attr in attributes],
        theta=attributes,
        fill = 'toself',
        name = '2018'
    )

    trace3 = go.Scatterpolar(
        r = [data17[attr].values[0] for attr in attributes],
        theta=attributes,
        fill = 'toself',
        name = '2017'
    )

    trace4 = go.Scatterpolar(
        r = [data16[attr].values[0] for attr in attributes],
        theta=attributes,
        fill = 'toself',
        name = '2016'
    )

    data = [trace0,trace1,trace2,trace3,trace4]

    layout= go.Layout(
        polar = dict(
            radialaxis = dict(
                visible=True,
                range=[0,100]
            )
        ),
        showlegend=True,
        title=f'Scatter Polar of {name} Performance from 2016-2020'

    )

    fig = go.Figure(data=data,layout=layout)
    fig.show()
```

### Pick Top 5 Players per Position
Creating a method to pick top 5 player based on a the player position and the player value in euro

```python
cols = ['short_name','age','nationality','club','overall','value_eur']
# attack = ['RW', 'LW', 'ST', 'CF', 'LS', 'RS', 'RF', 'LF']
def top_players(pos):
    sample = df[df['team_position'] == pos.upper()]
    return sample.sort_values(by='value_eur',ascending=False)[cols].head(5)
top_players('lw')
```
