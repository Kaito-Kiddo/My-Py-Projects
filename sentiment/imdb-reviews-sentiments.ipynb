## Task 3: Decoding the Reviews
___
Note: If you are starting the notebook from this task, you can run cells from all previous tasks in the kernel by going to the top menu and then selecting **Kernel > Restart and Run All**
___


```python
reverse_word_index = dict((value, key) for key, value in word_index.items())

def decode(review):
    text = ''
    for i in review:
        text += reverse_word_index[i]
        text += ' '
    return text
```

```python
decode(x_train[0])
```

## Task 6: Creating and Training the Model
___
Note: If you are starting the notebook from this task, you can run cells from all previous tasks in the kernel by going to the top menu and then selecting **Kernel > Restart and Run All**
___

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Embedding, Dense, GlobalAveragePooling1D

model = Sequential([
    Embedding(10000, 16),
    GlobalAveragePooling1D(),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['acc']
)

model.summary()
```

```python
from tensorflow.python.keras.callbacks import LambdaCallback

simple_logging = LambdaCallback(on_epoch_end = lambda e, l: print(e, end='.'))

E = 20

h = model.fit(
    x_train, y_train,
    validation_split = 0.2,
    epochs = E,
    callbacks = [simple_logging],
    verbose = False
)
```

## Task 7: Predictions and Evaluation

```python
print(h.history.keys())
```

```python
import matplotlib.pyplot as plt
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])

plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training and Validation Losses')
plt.xlabel('Epoch Number')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
plt.plot(h.history['acc'])
plt.plot(h.history['val_acc'])

plt.title('Model accuracy Progression During Training/Validation')
plt.ylabel('Training and Validation accuracy')
plt.xlabel('Epoch Number')
plt.legend(['Training accuracy', 'Validation accuracy'])
```

```python
loss , acc  = model.evaluate(x_test,y_test)
```

```python
print(f'Accuracy : {acc*100}')
```

```python
predicted = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
import numpy as np
import seaborn as sns
# print(confusion_matrix(y_test, np.round(predicted)))
cm = confusion_matrix(y_test, predicted.round())
sns.heatmap(cm, annot = True)
```
